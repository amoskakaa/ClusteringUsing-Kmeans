```python
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist
wcss = []
K = range(1,11)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
   # wcss.append(kmeans.inertia_)
    wcss.append( sum(np.min(cdist(data, kmeans.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0])

# Plot the elbow
plt.plot(K, wcss, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method for KMeans Clustering')
plt.show()
```

From the figure above, as the value of k increases, the centroids are closer to the clusters until some point where there is rapid change and this point is at k =3. Thus, the most suitable value of k would be 3.

```python
Km = KMeans(n_clusters=3)
Km.fit(data) 
print(Km.cluster_centers_)  
```

```python
Km.labels_
```

Use the labels to assign a point to its respective cluster.

```python
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('V2 AGAINST V1')
plt.scatter(X,Y, marker = '+', c= Km.labels_, s = 50, cmap = 'plasma' )
plt.scatter( 9.4780459,10.686052, s = 60, c='r', marker = '*', label ='Centroid A')
plt.scatter( 40.68362784,59.71589274, s = 60, c='black', marker = '*',label = 'Centroid B')
plt.scatter( 69.92418447,-10.11964119, s = 60, c='g', marker = '*',label = 'Centroid C')
plt.legend(loc='upper right',scatterpoints=1)

```
